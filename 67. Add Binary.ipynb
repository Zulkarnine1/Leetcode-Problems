```python
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        len_a = len(a)
        len_b = len(b)
        if(len_a>len_b):
            b = "0"*(len_a-len_b) + b
        if(len_b>len_a):
            a = "0"*(len_b-len_a) + a
            
 
        p = len(a)-1
        carry = 0
        ans = ""
        while(p>=0):
            val = (int(a[p])+int(b[p])+carry)%2
            ans = str(val)+ans
            if int(a[p])+int(b[p])+carry > 1:
                carry = 1
            else: 
                carry = 0
            p-=1
            
        if(carry==0):
            return ans
        else:
            return str(carry)+ans
        
```
