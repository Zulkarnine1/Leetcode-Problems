```python
class Solution:
    def maxOperations(self, nums, k):
        # Sort the array first
        nums = sorted(nums)
        # Set left, right and count
        left = 0
        right = len(nums)-1
        count = 0
        while left < right:
            sum = nums[left] + nums[right]
            if sum==k:
                left = left + 1
                right = right - 1
                count = count + 1
            elif sum>k:
                right = right - 1
            elif sum<k:
                left = left + 1
      


        return count
```

```python
s1 = Solution()
ans1 = s1.maxOperations([1,2,3,4],5)
ans2 = s1.maxOperations([3,1,3,4,3],6)
print(ans1)
print(ans2)
```
