```python
class Solution:
    def isValid(self, s: str) -> bool:
        
        opn = ("(","{","[")
        close = {
            ")":opn[0],
            "}":opn[1],
            "]":opn[2]
                }
        if len(s) == 1 or s[0] in close:
            return False
        
        
        my_stack = []
        
        for char in s:
            if char in opn:
                my_stack.append(char)
            else:
                if len(my_stack) == 0 and char in close:
                    return False
                if my_stack[len(my_stack)-1] == close[char]:
                    my_stack.pop()
                else:
                    return False
                
        if len(my_stack) == 0:
            return True
        else:
            return False
```
