```python
class Solution:
    def plusOne(self, digits):
        rem = 1
        for i in range(len(digits)-1,-1,-1):
            val = (digits[i]+rem)%10
            digits[i] = val
            if(val==0):
                rem=1
            else:
                rem = 0
                break
                
        if rem >0:
            return [1]+digits
        else:
            return digits
```
