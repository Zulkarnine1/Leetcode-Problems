```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        n_len = len(needle)
        hy_len = len(haystack)
        if n_len == 0:
            return 0
        
        for i in range(hy_len):
            if haystack[i] == needle[0]:
                crop = i+n_len
                if crop>hy_len:
                    return -1
                else:
                    if haystack[i:crop] == needle:
                        return i
        return -1
```
