```python
def twoSum(nums, target):
        reslist = []
        for i in range(len(nums)):
            for j in range(len(nums)):
                if(i!=j):
                    res = nums[i] + nums[j]
                    if(res==target):
                        reslist.append(i)
        return reslist
```

```python
twoSum([2,7,11,15],9)
```
