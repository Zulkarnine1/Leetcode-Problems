```python
class Solution:
    def convert(self, s, numRows):
        t = 0
        i = 0
        hash = {}
        s_len = len(s)
        grow = True
        while (i<s_len):   
            if t == 0: grow = True
            if t == numRows-1: grow = False
            if (t in hash): 
                hash[t] = hash[t] + s[i]
            else:
                hash[t] = s[i]
            i+=1
            if grow:t+=1 
            else: t-=1
        res = "".join(hash.values())
        return res
```

```python
# Test cases:
s1 = Solution()
print(s1.convert("PAYPALISHIRING",3))
print(s1.convert("PAYPALISHIRING",4))
print(s1.convert("A",4))
```
