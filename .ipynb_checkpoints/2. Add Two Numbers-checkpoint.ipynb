```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1: ListNode, l2: ListNode) -> ListNode:
    remainder = 0
    # Keeping the Head for returning linkedlist
    resultLL = ListNode()
    # Temp for moving from one node to another
    temp = resultLL
    while l1 or l2:
        if l1:
            val1 = l1.val
        else : 
            val1 = 0
        if l2:
            val2 = l2.val
        else : 
            val2 = 0
        sum = val1 + val2 + remainder
        remainder = int(sum/10)
        result = sum%10
        # Adding val to next node
        temp.next = ListNode(result)
        # Moving to next node
        temp = temp.next

        if l1:l1 = l1.next
        if l2:l2 = l2.next
    if remainder > 0 : temp.next = ListNode(remainder)

    return resultLL.next
```

```python
# Sample Linked Lists
list1_node1 = ListNode(2)
list1_node2 = ListNode(4)
list1_node1.next = list1_node2
list1_node3 = ListNode(3)
list1_node2.next = list1_node3

list2_node1 = ListNode(5)
list2_node2 = ListNode(6)
list2_node1.next = list2_node2
list2_node3 = ListNode(4)
list2_node2.next = list2_node3
```

```python
addTwoNumbers(list1_node1,list2_node1)
```
