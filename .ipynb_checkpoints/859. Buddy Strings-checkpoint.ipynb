```python

def buddyStrings( a: str, b: str) -> bool:
    lena = len(a)
    lenb = len(b)

    if(lena!=lenb):
        return False

    if(a==b):
        if(len(set(a))>=lena):
            return False
        else:
            return True

    diff = []

    for i in range(lena):
        if(a[i]!=b[i]):
            diff.append(i)

    if(len(diff)==2):
        return a[diff[0]] == b[diff[1]] and b[diff[0]] == a[diff[1]]
    else:
        return False
        
```

```python
buddyStrings("ab","ba")
```
