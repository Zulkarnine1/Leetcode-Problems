```python
class Solution:
    def interpret(self, command):
        strlist = []
        for i in range(0, len(command)):
            firstChar = command[i]
            if(i+1<len(command)):
                secondChar = command[i+1]
                if(firstChar=="G"):
                    strlist.append("G")
                if(firstChar=="(" and secondChar==")"):
                    strlist.append("o")
                    i=i+2
                if(firstChar=="(" and secondChar=="a"):
                    strlist.append("al")
                    i=i+4
            elif(firstChar=="G"):
                strlist.append("G")
        result = ''.join(map(str, strlist))
        return result
```

```python
s1 = Solution()
ans1 = s1.interpret("G()(al)")
ans2 = s1.interpret("G()()()()(al)")
ans3 = s1.interpret("(al)G(al)()()G")
print(ans1)
print(ans2)
print(ans3)
```
