```python
class Solution:
    def longestCommonPrefix(self, strs) -> str:
        # Check for one element array
        if len(strs) == 1:
            return strs[0]
        # Find the shortest word
        shortest_len = len(strs[0])
        shortest_word = strs[0]
        
        for word in strs:
            temp = len(word)
            if temp<shortest_len:
                shortest_len = temp
                shortest_word = word
                
        
        # For each word check the matching inital indexes 
        broken = False
        index = 0
        while (not broken):
            for word in strs:
                if index>= shortest_len:
                    broken = True
                    break
                if word[index] != shortest_word[index]:
                    if index == 0:
                        return ""
                    broken = True
                    break
            index += 1
        return shortest_word[:index-1]
            
                
        
```

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        l = list(zip(*strs))
        prefix = ""
        for i in l:
            if len(set(i))==1:
                prefix += i[0]
            else:
                break
        return prefix
```
