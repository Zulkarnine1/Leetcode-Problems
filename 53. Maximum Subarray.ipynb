```python
class Solution:
    def maxSubArray(self, nums) -> int:
        cm = nums[0]
        gm = nums[0]
        
        for i in range(1,len(nums)):
            if(cm+nums[i]>=nums[i]):
                cm = cm+nums[i]
            else:
                cm = nums[i]
                
            if(cm>gm):
                gm=cm
                
        return gm
        
```
