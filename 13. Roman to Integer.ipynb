```python
# A solution from another person which is much better

class Solution:
    def romanToInt(self, s: str) -> int:
        req = 0
        bas = {"I":1, "V":5, "X":10, "L": 50, "C": 100, "D": 500, "M": 1000}
        com = {"IX": 9, "IV": 4, "XL": 40, "XC": 90, "CD": 400, "CM": 900}
        
        i = 0
        while i<len(s):
            if i+1<len(s) and com.get(s[i]+s[i+1], False):
                req+=com.get(s[i]+s[i+1])
                i+=2
            else:
                req+=bas.get(s[i])
                i+=1
        return req
```
