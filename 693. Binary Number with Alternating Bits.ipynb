```python
class Solution:
    def majorityElement(self, nums) -> int:
        
        lim = len(nums)/2
        
        myhash = {}
        for i in range(len(nums)):
            if (nums[i] in myhash):
                myhash[nums[i]] += 1
            else:
                myhash[nums[i]] = 1
                
        for key,val in myhash.items():
            if(val>lim):
                return key
 
        return key
                
            
        
        
```
