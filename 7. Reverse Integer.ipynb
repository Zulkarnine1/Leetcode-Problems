```python
# Recursive method

class Solution:
    def reverse(self, x: int) -> int:
        x = str(x)
        if "-" not in x: 
            sol = int(self.recurrev(x))
        else:
            sol = -1*int(self.recurrev(x[1:]))
        
        if ((2**31)-1)<= sol or sol <= ((-2)**31):
            return 0
        else:
            return sol
        
        
    def recurrev(self,x):
        if len(x) <= 1:
            return x
        else:
            return self.recurrev(x[1:])+x[0]
        
```

```python
# Iterative with type conversion method

class Solution:
    def reverse(self, x: int) -> int:
        x = str(x)
        if "-" not in x: 
            sol = int(self.iterrev(x))
        else:
            sol = -1*int(self.iterrev(x[1:]))
        
        if ((2**31)-1)<= sol or sol <= ((-2)**31):
            return 0
        else:
            return sol
        
        
    def iterrev(self,x):
        length = len(x)
        if length <= 1:
            return x
        else:
            temp = ""
            for i in range(length-1,-1,-1):
                temp += x[i]
            
            return temp
            
            
```
