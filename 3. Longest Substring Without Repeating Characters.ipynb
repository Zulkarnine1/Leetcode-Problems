```python
class Solution:
    def lengthOfLongestSubstring(self, s):
        # Init mylist with 0 to have minimum 0 output 
        my_list = [0]
        length = len(s)
        # Counter max_count
        max_count = 0
        # Interate through the string starting from each char and moving on to the next
        for i in range(0,length):
            # Break if max already greater than the rest of the string
            if((length-i)<max_count):break
            string = ""
            if i+1<=length:
                # Iterate through the string from this char at i index
                for j in range(i,length):
                    # If char not in string add to string
                    if(s[j] not in string):
                        string += s[j]
                        # If second last char, push string len to list and update max_count
                        if(length==j+1):
                            s_length = len(string)
                            my_list.append(s_length)
                            if(s_length>max_count): max_count = s_length
                    # If char in string already update max_count and break loop
                    else:
                        s_length = len(string)
                        my_list.append(s_length)
                        if(s_length>max_count): max_count = s_length
                        break
        # Return max_count
        return max(my_list)
    # ====================== ====================
    # ====================== ====================
    #                 Faster Method
    # ====================== ====================
    # ====================== ====================
    
    def lengthOfLongestSubstring1(self, s):
        start = maxLength = 0
        usedChar = {}
        
        for i in range(len(s)):
            # In order to change the starting point
            if s[i] in usedChar and start <= usedChar[s[i]]:
                start = usedChar[s[i]] + 1
            else:
                # In order to change the max
                maxLength = max(maxLength, i - start + 1)
            # Using hastable to record the entries of each char
            usedChar[s[i]] = i

        return maxLength
```

```python
def lengthOfLongestSubstring(s):
    dic, res, start, = {}, 0, 0
    for i, ch in enumerate(s):
        if ch in dic:
            res = max(res, i-start) # update the res
            start = max(start, dic[ch]+1)  # here should be careful, like "abba"
        dic[ch] = i
    return max(res, len(s)-start)  # return should consider the last non-repeated substring
```
