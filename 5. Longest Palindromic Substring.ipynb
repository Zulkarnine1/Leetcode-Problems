```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        s_len = len(s)
        max_count = 0
        max_s = ""
        
        for i in range(s_len):
            l = r = i
            while (l>=0 and r< s_len and s[l]==s[r]):          
                if r-l+1 > max_count:
                    max_s = s[l:r+1]
                    max_count = r-l+1
                l-=1
                r+=1
          
    
        for i in range(s_len):
            l =  i
            r = i + 1
            while (l>=0 and r< s_len and s[l]==s[r]):          
                if r-l+1 > max_count:
                    max_s = s[l:r+1]
                    max_count = r-l+1
                l-=1
                r+=1


          
        
        return max_s
```

```python
s1 = Solution()
print(s1.longestPalindrome("babad"))
print(s1.longestPalindrome("cbbd"))
print(s1.longestPalindrome("a"))
print(s1.longestPalindrome("ac"))
print(s1.longestPalindrome("bb"))
print(s1.longestPalindrome("aaaa"))
print(s1.longestPalindrome("aacabdkacaa"))
```
