```python
# Solution with pallindrome iteration reversal 

class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x<0:
            return False
        else:
            
            def rev(n):
                res = ""
                for i in range(len(n)-1,-1,-1):
                    res += n[i]

                return int(res)
            
            return x == rev(str(x))
```

```python
# Solution with pallindrome iteration arithmetic reversal 

class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x<0:
            return False
        else:
            
            def rev(n):
                res = 0
                while(n>=1):
                    res = (res*10)+(n%10)
                    n = n//10
                    
                return res
            
            return x == rev(x)
        
```
