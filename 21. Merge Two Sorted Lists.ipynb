```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        res = ListNode()
        cur = res
        p1 = l1
        p2 = l2
        
        while(p1 and p2):
            if(p1.val<p2.val):
                newNode = ListNode(p1.val)
                cur.next = newNode
                cur = newNode
                p1 = p1.next
            else:
                newNode = ListNode(p2.val)
                cur.next = newNode
                cur = newNode
                p2 = p2.next
        remainder = None
        if p1:
            remainder = p1
        if p2:
            remainder = p2
            
        while remainder:
            newNode = ListNode(remainder.val)
            cur.next = newNode
            cur = newNode
            remainder = remainder.next
            
        return res.next
                
        
```
