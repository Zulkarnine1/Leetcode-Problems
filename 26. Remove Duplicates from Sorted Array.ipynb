```python
class Solution:
    def removeDuplicates(self, nums) -> int:
        for i in range(len(nums)):
            j = i+1
            while(j<len(nums)):
                if nums[j]==None or nums[j] > nums[i] :
                    break
                if nums[i] == nums[j]:
                    nums.pop(j)
                    j-=1
                j+=1
                
                    
        
        return len(nums)
        
        
```
