```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        com_arr = sorted(nums1 + nums2)
        median = None
        mid = len(com_arr)//2     
        if (len(com_arr)%2==0):        
            median = (com_arr[mid-1] + com_arr[mid])/2
        else :       
            median = com_arr[mid]
        return median
```

```python
# Test cases:
s1 = Solution()
print(s1.findMedianSortedArrays([1,3],[2]))
print(s1.findMedianSortedArrays([1,2],[3,4]))
print(s1.findMedianSortedArrays([0,0],[0,0]))
print(s1.findMedianSortedArrays([],[1]))
print(s1.findMedianSortedArrays([2],[]))
```
